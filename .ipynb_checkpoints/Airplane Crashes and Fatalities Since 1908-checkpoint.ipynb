## 1.1 basic analyze

```python
crash_df.info()
```

```python
# add a new column "Year"
years = []
for year in crash_df['Date']:
    years.append(year[6:])
crash_df['Year'] = years
crash_df.head()
```

```python
yearly_df = crash_df[['Year','Aboard','Fatalities','Ground']].groupby('Year').agg(['sum','count']) #  do not forget include groupby column
yearly_df.head()
```

```python
yearly_df['FAP'] = yearly_df.Fatalities['sum']/yearly_df.Aboard['sum']    # call multi-level column by using a.b['c']  !!!!
yearly_df['GAP'] = yearly_df.Ground['sum']/yearly_df.Aboard['sum']
yearly_df.tail()
```

```python
print "The total number of Airplane Crashes is %i since 1908 to 2009." %sum(yearly_df.Aboard['count'])
print "The total number of Aboard is %i since 1908 to 2009." %sum(yearly_df.Aboard['sum'])
print "The total number of Fatalities is %i since 1908 to 2009." %sum(yearly_df.Fatalities['sum'])
print "The total number of Ground is %i since 1908 to 2009." %sum(yearly_df.Ground['sum'])
print "\n"
print "The percentage of fatalities is %.2f%% since 1908 to 2009." %72.97
```

```python
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(4,1,1);ax1.plot(yearly_df.Aboard['count'],'g-o');plt.xlim([1908,2005]);ax1.set_xlabel('YEAR');ax1.set_title('Crash Numbers / YEAR')
# 'g-0'   equal to : color='g',linestyle='-',marker='o'
ax2 = fig.add_subplot(4,1,2);ax2.plot(yearly_df.Aboard['sum'],'k-x');plt.xlim([1908,2005]);ax2.set_xlabel('YEAR');ax2.set_title('Aboard Numbers / YEAR')
ax3 = fig.add_subplot(4,1,3);ax3.plot(yearly_df.Fatalities['sum'],'r-x');plt.xlim([1908,2005]);ax3.set_xlabel('YEAR');ax3.set_title('Fatalities Numbers / YEAR')
ax4 = fig.add_subplot(4,1,4);ax4.plot(yearly_df.Ground['sum'],'b-x');plt.xlim([1908,2005]);ax4.set_xlabel('YEAR');ax4.set_title('Ground Numbers / YEAR')
```

```python
fig,axes = plt.subplots(2,1,figsize=(15,10),sharex=True,sharey=True)
plt.subplots_adjust(wspace=0.5,hspace=0.5)
ax5 = fig.add_subplot(2,1,1); ax5.plot(yearly_df.FAP,'r-o');plt.xlim([1908,2005]);ax5.set_xlabel('YEAR');ax5.set_title('Fatalities Percentage Yearly')
ax6 = fig.add_subplot(2,1,2); ax6.plot(yearly_df.GAP,'g-p');plt.xlim([1908,2005]);ax6.set_xlabel('YEAR');ax6.set_title('Ground Percentage Yearly')
#ax5.text(1992,0.32,'1999(0.35)',family='monospace',fontsize=10)
lowest_year = [(1999,0.35),(1908,0.50),(1990,0.55)]
for year, label in lowest_year:
    ax5.annotate(label,xy=(year,label),arrowprops=dict(facecolor='blue'))
ax6.annotate(2.67,xy=(2001,2.67),arrowprops=dict(facecolor='black'))
```
