```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import Series,DataFrame
```

```python
url = 'http://www.sundxswk.com/phb/119.html'
```

```python
result = requests.get(url)
c = result.content
soup = BeautifulSoup(c,"lxml")
```

```python
summary = soup.find('div',{'class':'w960 center clear mt2'})
tables = summary.find_all('table')
```

```python
data = []
rows = tables[0].findAll('tr')
for tr in rows:
    cols = tr.findAll('td')
    tempRowData = []
    for td in cols:
        text = td.find(text=True) 
        tempRowData.append(text)
    print tempRowData
    data.append(tempRowData)
```

```python
df1 = DataFrame(data)
df2 = df1.drop(0)
df2.columns = ['排名','去年排名','公司名称','营业收入','利润','国家']
df1.tail(10)
```

```python
df2.count()
```

```python
df3 = df2[['利润','国家']].groupby(['国家']).agg(['sum','count'])
df3
```
